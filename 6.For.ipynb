# Nested For Loop

```python
a = ['red','green','yellow']
b = ['cherry','apple','banana']

for x in a:
    for y in b:
        print(x,y)
```

```python
a = [5,6]
b = [1,2,3,4,5]

for x in a:
    for y in b:
        print("{} x {} = {}".format(x,y,x*y))
    print("\n***********\n")
```
