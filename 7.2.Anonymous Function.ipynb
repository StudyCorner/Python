# Use of Lambda Function in Python

```python
my_lis = [1,2,3,4,5]

even_ = list(filter(lambda x:(x%2==0),my_lis))

even_
```

```python
my_lis = [1,2,3,4,5]

twice_ = list(map(lambda x:(x*2),my_lis))

twice_
```

```python
my_lis = [1,2,3,4,5]

even_ = list(map(lambda x:(x%2==0),my_lis))

even_
```

```python
import functools
```

```python
lis = [1,3,5,6,2]

functools.reduce(lambda a,b:a+b,lis)
```

```python
lis = [1,3,5,6,2]

functools.reduce(lambda a,b:a-b,lis)
```

```python
lis = [1,3,5,6,2]

functools.reduce(lambda a,b:a if a>b else b,lis)
```
