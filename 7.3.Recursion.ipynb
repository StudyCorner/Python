

> A function that calls itself is called as Recursion



5! = 5 x 4 x 3 x 2 x 1 = 120

```python
def factorial(x):
    if x ==1:
        return  1
    else:
        return (x*factorial(x-1))

n = 3
print("The Factorial of ",n , " : ",factorial(n))
```

```python
5*factorial(4)
5*(4*factorial(3))
5*(4*(3*factorial(2)))
5*(4(*3(*2*factorial(1))))
5*(4*(3*(2*(1))))
```

Nested Functions

```python
def outer():
  print("Function 1 called")
  def inner():
    print("Function 2 called")
    print("Function 2 ---> Done")
    def inner1():
      print("Function 3 called & done")
    inner1()
  inner()
  print("Function 1 ---> Done")
```
