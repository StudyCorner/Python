A Variable scope specifies the region where we can access a variable

1 . Local Variable

2. Global Variable

3. Non Local variable

**Local Variable**

```python
def fun():
  x_ = "local"
  print("Inside Func ",x_)

#x = "global"
fun()
print("Outside Func ",x_)
```

```python
def total_cost():  # function header
  pencil = 5
  no_pencils = 10
  tc = pencil*no_pencils
  print("Total Cost : ",tc)
```

```python
total_cost()
```

```python
# access local variable locally

def abc():
    a = 10;
    print(a)
    
abc()
```

```python
b = 'global'  # global variable
def abc():
    print(b)  # accessing  global variable locally
abc()
print(b)  # global access
```

```python
light = "Sun light"  # global
def home():
  light = "Tube light"  # local variable
  print(light)
```

```python
home()
print(light)
```

```python
b = 'global'
def abc():
    print(b)  # accessing  global variable locally
abc()
print(b)   # accessing global variable globally
```

```python
def abc():
    c= 'local'
    print(c)   # accessing  locally variable locally
abc()
print(c)  # accessing  local variable globally
```

```python
d = 'global'
def fun():
    d = d +' local' # modifying global var locally is not possible
    print(d)

fun()
```

```python
d = 'global'
def fun():
    global d
    d = d +' local' # modifying global var locally is  possible by using global keyword
    print(d)

fun()
print(d)
```
