```python
x = 20
def outer():
  print("In Enclosing function Scope ....")
  x = 10  # non local scope
  def inner():
    print("Nested Function or inner")
    nonlocal x
    x = x + 5
    print("x in enclosing scope accesed via nested --> ",x)
  inner()
  print("End of Outer")
```
