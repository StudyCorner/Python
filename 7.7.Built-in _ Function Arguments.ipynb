```python
def greet(name,msg='Good Morning !!!'):
    print("Hello ", name + " , "+msg)
```

```python
greet('Sushma')
```

```python
greet('Raam','Have a nice day')
```

```python
def greet(name='ahmed',msg):
    print("Hello ", name + " , "+msg)
```

```python
def greet(name ='Narmadha',msg='Good Morning !!!'):
    print("Hello ", name + " , "+msg)
```

```python
def greet(*names):
    for name in names:
        print("Hello ",name)
    print(type(names))
```

```python
greet('neema','reema','emma','semma','leema')
```

```python
def greet(**wish):
    print('Name : ',wish['name'])
    print('Message  : ',wish['msg'])
    print('Last Name : ',wish.get('lname'))
    print(type(wish))
```

```python
greet(name = 'Jhon',msg = 'Congratulations',lname = 'Refenses')
```
